```python
import pandas as pd
import numpy as np
import re
```

```python
billion=pd.read_csv('./../CSV/RawCSV/BillionairesIndex.csv')
GDP=pd.read_csv('./../CSV/RawCSV/Countries GDP 1960-2020.csv')
ICT=pd.read_csv('./../CSV/RawCSV/ICTSkills_Kaggle.csv', encoding='latin1')
```

```python
billion['Total_Net_Worth']=billion['Total_Net_Worth'].apply(lambda x: re.sub(r'\D', '', x))
billion['Total_Net_Worth']=billion['Total_Net_Worth'].astype(int)
```

```python
len(billion['Country'].unique())
```

```python
billion[['Name', 'Total_Net_Worth', 'Country','Industry']].groupby('Country').agg({'Total_Net_Worth': 'sum', 'Name': lambda x: ', '.join(x),'Industry': lambda x: list(set((', '.join(x)).split(',')))}).sort_values(by='Total_Net_Worth',ascending=False)
```

```python
billion[['Name', 'Total_Net_Worth', 'Industry']].groupby('Industry').agg({'Total_Net_Worth': 'sum', 'Name': lambda x: ', '.join(x)}).sort_values(by='Total_Net_Worth',ascending=False)
```

```python
choroplethmap=billion[['Name', 'Total_Net_Worth', 'Country','Industry']].groupby('Country').agg({'Total_Net_Worth': 'sum', 'Name': lambda x: ', '.join(x),'Industry': lambda x: list(set((', '.join(x)).split(',')))}).sort_values(by='Total_Net_Worth',ascending=False)
choroplethmap.to_csv("./../CSV/ProcessedCSV/totalnetworth.csv")
```
